```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
import pandas as pd
```

```python
nba_players = pd.read_csv('./players_stats.csv')
nba_players = nba_players.drop(['Name', 'Games Played', 'Birth_Place', 'Birthdate', 'Collage', 'Team', 'Experience'], axis=1)
nba_players = nba_players.reset_index(drop=True)
```

```python
idx = 0
badidx = []
for stat in nba_players:
    for i in range(nba_players.shape[0]):
        if type(nba_players[stat][i])==type(1.1) and math.isnan(nba_players[stat][i]):
            badidx.append(i)
            
badidx = set(badidx)
```

```python
nba_players = nba_players.drop(badidx)
nba_players = nba_players.reset_index(drop=True)
nba_players
```

```python
nba_players.to_csv('NBA_clean.csv')
```
